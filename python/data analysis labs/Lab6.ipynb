```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

# Задания

**Задание 1.**  Используя данные выше:
1. Создайте объект figure с именем fig, используя plt.figure()
2. Используйте add_axes, чтобы добавить ось к фигуре в [0,0,1,1].
3. Нанесите (x, y) на эти оси и установите метки и заголовки так, чтобы они соответствовали приведенному ниже графику:


```python
fig = plt.figure()
fig.add_axes([0,0,1,1])
plt.plot(x, 2*x)

plt.xlabel('x')
plt.locator_params(axis = 'x', nbins = 6)
plt.xlim(0,100)

plt.ylabel('y')
plt.locator_params(axis = 'y', nbins = 5)
plt.ylim(0,200)

plt.title("title")
plt.show()
```

**Задание 2.** Используя данные нарисуйте график, чтобы он соответствовал приведенному ниже:


```python
fig = plt.figure()
fig.add_axes([0,0,1,1])
plt.plot(x, 2*x)

plt.xlabel('x')
plt.xlim(0,100)
plt.locator_params(axis = 'x', nbins = 6)

plt.ylabel('y')
plt.ylim(0,200)
plt.locator_params(axis = 'y', nbins = 5)

fig.add_subplot(445)
plt.plot(x, 2*x)

plt.xlabel('x')
plt.xlim(0,100)
plt.locator_params(axis = 'x', nbins = 6)

plt.ylabel('y')
plt.ylim(0,200)
plt.locator_params(axis = 'y', nbins = 5)

plt.show()
```

**Задание 3.**  Используя данные нарисуйте график, чтобы он соответствовал приведенному ниже:


```python
fig = plt.figure()
fig.add_axes([0,0,1,1])
plt.plot(x, x**2)

plt.xlabel('X')
plt.xlim(0, 100)
plt.locator_params(axis = 'x', nbins = 6)

plt.ylabel('Z')
plt.ylim(0, 10000)
plt.locator_params(axis = 'y', nbins = 5)

fig.add_subplot(221)
plt.plot(x, x*2)
plt.title("zoom")

plt.xlabel('X')
plt.xlim(20.0, 22.0)
plt.locator_params(axis = 'x', nbins = 5)

plt.ylabel('Y')
plt.ylim(30, 50)
plt.locator_params(axis = 'y', nbins = 5)

plt.show()
```

**Задание 4.**  Используя данные нарисуйте график, чтобы он соответствовал приведенному ниже (здесь необходимо использовать `suplots`):


```python
fig = plt.figure()

plt.subplot(121)
plt.plot(x, 2*x, 'b--', linewidth = 3)

plt.xlim(0,100)
plt.locator_params(axis = 'x', nbins = 6)

plt.ylim(0,200)
plt.locator_params(axis = 'y', nbins = 5)

plt.subplot(122)
plt.plot(x, x**2, 'r-', linewidth = 3)

plt.xlim(0,100)
plt.locator_params(axis = 'x', nbins = 6)

plt.ylim(0, 10000)
plt.locator_params(axis = 'y', nbins = 6)

plt.show()
```
